```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
views=pd.read_csv(r'C:\Users\SYED\Desktop\Aedifico\24-10-2019\CSV Files\views_test.csv')
```

```python
views.head(7)
```

```python
x_axis=views['days_name']
y_views=views['youtube_views']
f_views=views['facebook_views']
t_views=views['twitter_views']
```

```python
#ploting all lines on a single graph

plt.style.use('ggplot')
plt.figure(figsize=(12,8))
plt.plot(x_axis,y_views, color='r',marker='o',markerfacecolor='r',label='Youtube Views')
for i,text in enumerate(y_views):
    plt.annotate(text,(x_axis[i],y_views[i]),xytext=(x_axis[i],y_views[i]+10),color='r',fontsize=13,rotation=20)

plt.plot(x_axis,f_views,linestyle='-.',color='b',marker='o',markerfacecolor='b',label='Facebook Views')
for i,txt in enumerate(f_views):
    plt.annotate(txt,(x_axis[i],f_views[i]),xytext=(x_axis[i],f_views[i]+10),color='b',fontsize=13,rotation=20)
    
plt.plot(x_axis,t_views,linestyle=':',color='g',marker='o',markerfacecolor='g',label='Twitter Views')
for i,tx in enumerate(t_views):
    plt.annotate(tx,(x_axis[i],t_views[i]),xytext=(x_axis[i],t_views[i]+10),color='g',fontsize=13,rotation=20)

plt.ylim(100,700)
plt.xticks(x_axis,rotation=20)
plt.xlabel('Week Days',size=18)
plt.ylabel('Views',size=18)
plt.title('Views On Social Channel on Week Days',size=20)
plt.legend(loc=3,fontsize=12)
plt.grid(True,linewidth=1,linestyle='-.')
plt.show()
```

<b>creating subplot</b>


```python
#plotting subplots

plt.style.use('ggplot')
plt.figure(figsize=(10,20))

fig,(ax1,ax2,ax3)=plt.subplots(nrows=3,ncols=1,sharex=True)

#subplot1 on ax1
ax1.plot(x_axis,y_views, color='r',marker='o',markerfacecolor='r',label='Youtube Views')
for i,text in enumerate(y_views):
    ax1.annotate(text,(x_axis[i],y_views[i]),xytext=(x_axis[i],y_views[i]+10),color='r',fontsize=8,rotation=20)
ax1.set_title('Views On Social Channel on Week Days',size=12)
ax1.legend(loc=3)
ax1.set_ylabel('Views',size=10)
ax1.grid(True,linewidth=1,linestyle='-.')

#subplot2 on ax2
ax2.plot(x_axis,f_views,linestyle='-.',color='b',marker='o',markerfacecolor='b',label='Facebook Views')
for i,txt in enumerate(f_views):
    ax2.annotate(txt,(x_axis[i],f_views[i]),xytext=(x_axis[i],f_views[i]+10),color='b',fontsize=8,rotation=20)
ax2.legend(loc=3)
ax2.set_ylabel('Views',size=10)
ax2.grid(True,linewidth=1,linestyle='-.')

#subplot3 on ax3
ax3.plot(x_axis,t_views,linestyle=':',color='g',marker='o',markerfacecolor='g',label='Twitter Views')
for i,tx in enumerate(t_views):
    ax3.annotate(tx,(x_axis[i],t_views[i]),xytext=(x_axis[i],t_views[i]+10),color='g',fontsize=8,rotation=20)
ax3.legend(loc=3)
ax3.set_ylabel('Views',size=10)
ax3.grid(True,linewidth=1,linestyle='-.')


plt.ylim(100,700)
plt.xticks(x_axis,size=8,rotation=20)
plt.xlabel('Week Days',size=13)


plt.tight_layout()
plt.show()
```
