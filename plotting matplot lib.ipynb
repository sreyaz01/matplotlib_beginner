```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv(r'C:\Users\SYED\Desktop\Aedifico\24-10-2019\CSV Files\car.csv')
```

```python
df.head()
```

```python
df_new=df.head(20)
```

```python
df_new
```

```python
x_axis=df_new['Year']
```

```python
x_axis
```

```python
y_axis=df_new['MSRP']
```

```python
y_axis
```

```python
plt.style.use('ggplot')
plt.figure(figsize=(20,10))
plt.scatter(x_axis,y_axis,100,label='Price')
plt.xticks(range(1992,2014))
plt.xlabel('Year',size=20)
plt.ylabel('Price',size=20)
plt.title('Car Price During Year 1992-2013',size=30)
plt.grid('True',color='g',linewidth=1,linestyle='-.')
for i,text in enumerate (y_axis):
    plt.annotate(text,(x_axis[i],y_axis[i]),xytext=(x_axis[i]+.03,y_axis[i]),fontsize=12,rotation=20)
plt.legend(loc='upper left',fontsize='xx-large')

plt.show()
```
